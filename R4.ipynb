```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return x + y
```

```python
def runge_kutta_4th_order(x0, y0, h, x_end):
    x_values = [x0]
    y_values = [y0]
    x = x0
    y = y0

    while x < x_end:
        k1 = h * f(x, y)
        k2 = h * f(x + h/2, y + k1/2)
        k3 = h * f(x + h/2, y + k2/2)
        k4 = h * f(x + h, y + k3)

        y = y + (1/6) * (k1 + 2*k2 + 2*k3 + k4)
        x = x + h

        x_values.append(x)
        y_values.append(y)

    return x_values, y_values
```

```python
x0 = 0
y0 = 1
h = 0.1
x_end = 1
```

```python
x_values, y_values = runge_kutta_4th_order(x0, y0, h, x_end)
```

```python
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.title('Solución usando el Método de Runge-Kutta de 4to Orden')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid(True)
plt.show()
```
